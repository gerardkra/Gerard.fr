### Task 2.3

Using turtle, write a function draw_polygon(sides) that takes an integer parameter sides.

The function draws a regular polygon with the given number of sides:
- if sides = 3, then it draws an equilateral triangle ;
- if sides = 4, then it draws a square ;
- if sides = 5, then it draws a pentagon ;
- if sides = 6, then it draws a hexagon ;
 and soon...

```python
%%writefile polygon.py
import turtle as t

def draw_polygon(sides: int, side_len: int = 120, color="darkgreen"):
    if sides < 3:
        raise ValueError("sides must be >= 3")
    pen = t.Turtle()
    pen.color(color)
    pen.pensize(3)
    angle = 360 / sides
    for _ in range(sides):
        pen.forward(side_len)
        pen.right(angle)

if __name__ == "__main__":
    t.setup(700, 700)
    t.bgcolor("ivory")
    draw_polygon(5)  # try 3, 4, 6...
    t.exitonclick()
```

```python
from turtle import *
import time

def rounded_left(angle, radius):
    circle(radius, angle)

def draw_rosette_slow(petals=36, R=120, r=18, big_arc=90, round_angle=120, line_width=2, color='purple', pause=0.02):
    speed(4)               # slower – change 1..10 (0 is instant)
    pensize(line_width)
    pencolor(color)
    hideturtle()
    rotate_angle = 360 / petals

    for _ in range(petals):
        circle(R, big_arc); time.sleep(pause)
        rounded_left(round_angle, r); time.sleep(pause)
        circle(R, big_arc); time.sleep(pause)
        rounded_left(rotate_angle, r); time.sleep(pause)

    done()

draw_rosette_slow(petals=36, pause=0.02)
```

- Change symmetry: Adjust the last angle and loop count.
- For example, rounded_left(6, r) with range(60) (since 60×6=360).
- Softer or sharper corners: Change r. Larger r = gentler curve, smaller r = tighter corner.
- Scale the figure: Change R to make the main arcs bigger or smaller.
- Color & thickness: Try pencolor('cyan'), pensize(3), or even colormode(255) with RGB.

```python
from turtle import *
import math

def rounded_left(angle, radius):
    circle(radius, angle)

def draw_rosette_colored(petals=72, R=120, r=18, big_arc=90, round_angle=120, line_width=2, filename='rosette.ps'):
    colormode(255)   # allow RGB 0..255
    speed(0)
    pensize(line_width)
    hideturtle()
    rotate_angle = 360 / petals

    def color_wheel(t):
        """t in [0,1) -> smooth RGB via sine waves."""
        # Nice smooth cycling using phase shifts
        r = int(127.5 + 127.5 * math.sin(2*math.pi*(t)))
        g = int(127.5 + 127.5 * math.sin(2*math.pi*(t) + 2*math.pi/3))
        b = int(127.5 + 127.5 * math.sin(2*math.pi*(t) + 4*math.pi/3))
        return (r, g, b)

    for i in range(petals):
        pencolor(color_wheel(i / petals))
        circle(R, big_arc)
        rounded_left(round_angle, r)
        circle(R, big_arc)
        rounded_left(rotate_angle, r)

    # Save as PostScript (vector). Convert to PNG with your favorite tool.
    screen = getscreen()
    canvas = screen.getcanvas()
    canvas.postscript(file=filename)
    print(f"Saved as {filename}")

    done()

draw_rosette_colored(petals=72, filename='rosette.ps')
```

```python
main_py = """\
import sys
import pygame

def main():
    # 1) Initialize pygame
    pygame.init()

    # 2) Create a 600x600 window
    screen = pygame.display.set_mode((600, 600))
    pygame.display.set_caption("Hangman - Task 3.3")

    # A clock to limit the frame rate (prevents CPU from spinning at 100%)
    clock = pygame.time.Clock()

    # 3) Main loop
    running = True
    while running:
        # 4) Event handling
        for event in pygame.event.get():
            # Close when the window's X button is clicked
            if event.type == pygame.QUIT:
                running = False

            # Optional: close on Esc key
            elif event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE:
                running = False

        # 5) (Optional) Draw / update screen content
        screen.fill((240, 240, 240))  # light gray background

        # 6) Push the frame to the screen
        pygame.display.flip()

        # 7) Keep the loop at ~60 FPS
        clock.tick(60)

    # 8) Cleanup
    pygame.quit()
    sys.exit()

if __name__ == "__main__":
    main()
"""

with open("hangman/main.py", "w", encoding="utf-8") as f:
    f.write(main_py)

print("Updated hangman/main.py with Task 3.3 loop.")
```

```python
# Task 3.4 — Robust all-in-one cell (handles 403, has a local fallback)

import sys, subprocess, pkgutil, os, shutil, ssl, urllib.request
from pathlib import Path

# ---------- 0) Ensure deps ----------
def ensure_package(mod_name, pip_name=None):
    pip_name = pip_name or mod_name
    if pkgutil.find_loader(mod_name) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", pip_name, "--quiet"])

ensure_package("pygame")  # Pillow installed only if needed for fallback

# ---------- 1) Project structure ----------
root = Path("hangman")
assets_dir = root / "assets" / "images"
assets_dir.mkdir(parents=True, exist_ok=True)
bg_path = assets_dir / "background.jpg"

# ---------- 2) Try to download a web image (with headers) ----------
# Using browser-friendly direct image URLs. If they fail (403/offline), we’ll generate a local background.
# You can replace/add any direct image links you like.
candidate_urls = [
    # Unsplash raw images (commonly accessible via direct links)
    # Source pages for these images are on Unsplash (free to use); replace with your favorites if you want.
    "https://images.unsplash.com/photo-1501785888041-af3ef285b470?auto=format&fit=crop&w=1200&q=60",
    "https://images.unsplash.com/photo-1500530855697-b586d89ba3ee?auto=format&fit=crop&w=1200&q=60",
]

def try_download(url, dest_path: Path) -> bool:
    try:
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                                 "AppleWebKit/537.36 (KHTML, like Gecko) "
                                 "Chrome/122.0.0.0 Safari/537.36"}
        req = urllib.request.Request(url, headers=headers)
        context = ssl.create_default_context()
        with urllib.request.urlopen(req, context=context, timeout=30) as resp, open(dest_path, "wb") as out:
            shutil.copyfileobj(resp, out)
        return True
    except Exception as e:
        # Uncomment for debugging:
        # print(f"Download failed for {url}: {e}")
        return False

downloaded = False
if not bg_path.exists():
    for u in candidate_urls:
        if try_download(u, bg_path):
            downloaded = True
            break

# ---------- 3) Fallback: generate a sky gradient if download blocked ----------
if not downloaded and not bg_path.exists():
    # Install Pillow only if we need the fallback
    ensure_package("PIL", "Pillow")
    from PIL import Image

    W, H = 1200, 1200  # generate big and we will scale in pygame anyway
    top = (66, 135, 245)    # sky blue
    bottom = (220, 240, 255)
    img = Image.new("RGB", (W, H))
    # Vertical gradient
    for y in range(H):
        t = y / (H - 1)
        r = int(top[0]*(1-t) + bottom[0]*t)
        g = int(top[1]*(1-t) + bottom[1]*t)
        b = int(top[2]*(1-t) + bottom[2]*t)
        img.putpixel((0, y), (r, g, b))
    # Expand single-column gradient across width efficiently
    img = img.resize((W, H))
    img.save(bg_path, format="JPEG", quality=92)

# ---------- 4) Write main.py ----------
main_code = r'''# hangman/main.py
# Background image is stored at assets/images/background.jpg
# If you want to use another image, just replace that file.

import sys, os
import pygame

WIDTH, HEIGHT = 600, 600
BG_RELATIVE_PATH = os.path.join("assets", "images", "background.jpg")

def main():
    # Init pygame
    os.environ["PYGAME_HIDE_SUPPORT_PROMPT"] = "1"
    pygame.init()
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Hangman - Background")

    clock = pygame.time.Clock()

    # Resolve and verify background path
    here = os.path.dirname(__file__)
    bg_path = os.path.join(here, BG_RELATIVE_PATH)
    if not os.path.exists(bg_path):
        raise FileNotFoundError(f"Background image not found: {bg_path}")

    # Load and scale background once
    bg = pygame.image.load(bg_path).convert()
    bg = pygame.transform.smoothscale(bg, (WIDTH, HEIGHT))

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE:
                running = False

        screen.blit(bg, (0, 0))
        pygame.display.flip()
        clock.tick(60)

    pygame.quit()
    sys.exit()

if __name__ == "__main__":
    main()
'''

(root / "main.py").write_text(main_code, encoding="utf-8")

print("✅ Setup complete. Launching the game window...")
print("   Close the window (or press Esc) to return control to the notebook.")

# ---------- 5) Run the program ----------
subprocess.run([sys.executable, str(root / "main.py")])
```

### Task 3.5

```python
# Task 3.5 — Add a stickman drawing function to main.py and run it (single cell)

import sys, subprocess, pkgutil, os
from pathlib import Path

# Ensure pygame is installed
if pkgutil.find_loader("pygame") is None:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "pygame", "--quiet"])

root = Path("hangman")
assets_dir = root / "assets" / "images"
assets_dir.mkdir(parents=True, exist_ok=True)  # keep structure consistent
bg_path = assets_dir / "background.jpg"        # use if it exists from Task 3.4

main_code = r'''# hangman/main.py
import sys, os
import pygame

WIDTH, HEIGHT = 600, 600
BG_RELATIVE_PATH = os.path.join("assets", "images", "background.jpg")

def draw_stickman(surf, center, scale=1.0, color=(20, 20, 20), thickness=4):
    """
    Draw a simple stickman.
    center: (x, y) is the head center position.
    scale:  overall size multiplier (1.0 is default).
    color:  stroke color for outlines/lines.
    """
    x, y = center
    # --- Scaled dimensions ---
    head_r   = int(12 * scale)
    neck_y   = y + head_r
    body_len = int(40 * scale)
    hip_y    = neck_y + body_len
    arm_off  = int(10 * scale)   # shoulder drop below neck
    arm_len  = int(30 * scale)
    leg_len  = int(35 * scale)
    foot_dx  = int(18 * scale)

    # --- Head (circle outline) ---
    pygame.draw.circle(surf, color, (x, y), head_r, width=thickness)

    # --- Body ---
    pygame.draw.line(surf, color, (x, neck_y), (x, hip_y), thickness)

    # --- Arms (from shoulders) ---
    shoulder_y = neck_y + arm_off
    pygame.draw.line(surf, color, (x, shoulder_y), (x - arm_len, shoulder_y), thickness)  # left arm
    pygame.draw.line(surf, color, (x, shoulder_y), (x + arm_len, shoulder_y), thickness)  # right arm

    # --- Legs ---
    pygame.draw.line(surf, color, (x, hip_y), (x - foot_dx, hip_y + leg_len), thickness)  # left leg
    pygame.draw.line(surf, color, (x, hip_y), (x + foot_dx, hip_y + leg_len), thickness)  # right leg

    # (Optional) simple face details:
    eye_r = max(1, head_r // 6)
    eye_dx = max(2, head_r // 2)
    eye_y  = y - head_r // 4
    pygame.draw.circle(surf, color, (x - eye_dx, eye_y), eye_r)
    pygame.draw.circle(surf, color, (x + eye_dx, eye_y), eye_r)
    # smile
    smile_w = max(6, head_r)
    mouth_y = y + head_r // 4
    pygame.draw.arc(surf, color, (x - smile_w//2, mouth_y - eye_r, smile_w, head_r//2), 0.0, 3.1415, max(1, thickness-2))

def main():
    # Hide support prompt, init pygame
    os.environ["PYGAME_HIDE_SUPPORT_PROMPT"] = "1"
    pygame.init()

    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Hangman - Stickman")

    clock = pygame.time.Clock()

    # --- Load background if available, else just use a flat color ---
    here = os.path.dirname(__file__)
    bg_path = os.path.join(here, BG_RELATIVE_PATH)
    background = None
    if os.path.exists(bg_path):
        try:
            bg = pygame.image.load(bg_path).convert()
            background = pygame.transform.smoothscale(bg, (WIDTH, HEIGHT))
        except Exception:
            background = None

    running = True
    while running:
        # Events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE:
                running = False

        # Draw
        if background:
            screen.blit(background, (0, 0))
        else:
            screen.fill((235, 242, 255))  # soft sky as fallback

        # Draw the stickman centered nicely in the scene
        # Choose a scale that fits well in 600x600
        scale = 1.8
        head_x = WIDTH // 2
        head_y = HEIGHT // 2 - 80
        draw_stickman(screen, (head_x, head_y), scale=scale, color=(15, 15, 15), thickness=4)

        pygame.display.flip()
        clock.tick(60)

    pygame.quit()
    sys.exit()

if __name__ == "__main__":
    main()
'''

# Write/overwrite main.py
(root / "main.py").write_text(main_code, encoding="utf-8")

print("✅ main.py updated with draw_stickman(). Launching the window...")
print("   Close the window (or press Esc) to return to the notebook.")
subprocess.run([sys.executable, str(root / "main.py")])
```

```python
# Sierpiński triangle with turtle (single cell)
# Uses module-level turtle API (no Turtle instance)

from turtle import *

def serp_tri(side, level):
    """Draw an equilateral Sierpiński triangle with given side and recursion level.
       Assumes the turtle starts at the lower-left corner, heading to the right,
       and returns to the same pose (position + heading)."""
    if level == 0:
        # Draw a single filled-outline triangle (outline only here)
        for _ in range(3):
            fd(side)
            left(120)
    else:
        # Lower-left triangle
        serp_tri(side/2, level-1)
        # Move to lower-right
        fd(side/2)
        # Lower-right triangle
        serp_tri(side/2, level-1)
        # Return to lower-left
        bk(side/2)
        # Move to top position
        left(60); fd(side/2); right(60)
        # Top triangle
        serp_tri(side/2, level-1)
        # Return to lower-left original pose
        left(60); bk(side/2); right(60)

# ---- Setup & run ----
speed(0)            # 0 = fastest (no animation)
hideturtle()        # hide the turtle cursor
bgcolor('white')
color('black')

penup()
goto(-200, -200)    # starting corner
setheading(0)       # face right
pendown()

# Try level 6–8 depending on your machine
serp_tri(400, 6)

done()  # keep the window open
```
