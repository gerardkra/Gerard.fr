#### Découverte de quelques méthodes PySpark

```python
raw_data = sc.textFile('daily_show_guests.csv')
raw_data.take(10)
```

```python
daily_show = raw_data.map(lambda line: line.split(','))
daily_show.take(5)
```

```python
tally = daily_show.map(lambda x: (x[0], 1)).reduceByKey(lambda x,y: x+y)
```

```python
#on constate que le contenu de la variable tally ne s'affiche pas du fait du fonctionnement 'paresseux' des RDD
print(tally)
```

```python
#pour afficher le contenu de tally, il faut forcer l'exécution des transformations
tally.take(tally.count())
```

```python
def filter_year(line):
    # Write your logic here
    if line[0] == 'YEAR':
        return False
    else :
        return True

filtered_daily_show = daily_show.filter(lambda line: filter_year(line))
filtered_daily_show.take(5)
```

```python
filtered_daily_show.filter(lambda line: line[1] != '') \
                   .map(lambda line: (line[1].lower(), 1)) \
                   .reduceByKey(lambda x,y: x+y) \
                   .take(5)
```
